```python
def read_txt_file(filePath):
    try:
        with open(filePath, "r") as file:
            strings = file.readlines()
            print(strings)
    except Exception as ex:
        print("error catch : " , ex)
    
```

```python
filePath = 'fake_img_url_list.txt'

write_some_text(filePath,'Hello\nWorld!');
read_txt_file(filePath)
```

```python
## url (한글) decode
# sample_url = '%EC%8A%88%ED%8D%BC%EB%A9%A4%EB%B2%84%EC%8A%A4'
sample_url = '%EC%BA%A1%EC%B2%983'

# %EC%A0%9C%EB%AA%A9_%EC%97%86%EC%9D%8C
new_url = urllib.parse.unquote(sample_url)
print(new_url)
```

```python
## url (한글) encode
new_str = urllib.parse.quote_plus(str(new_url))
print(new_str)
```
